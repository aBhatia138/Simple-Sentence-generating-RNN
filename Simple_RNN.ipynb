```python
import numpy as np
from keras import layers,models
import tensorflow as tf
```

```python
text="I am Amoghaya Bhatia and i love coding"
chars=sorted(list(set(text)))
```

```python
char_to_idx={char:i for i,char in enumerate(chars)}
idx_to_char={i:char for i,char in enumerate(chars)}
```

```python
seq_length=3
sequences=[]
labels=[]
for i in range(len(text)-seq_length):
    seq=text[i:i+seq_length]#Sequence
    label=text[i+seq_length]#Character following the sequence
    sequences.append([char_to_idx[char] for char in seq])#List of lists that contains number equivalents from each seq. Number equivalents are derived from chars
    labels.append(char_to_idx[label])#List that contains character following each seq for each seq in sequences
X=np.array(sequences)
y=np.array(labels)
X
```

```python
from keras.utils import to_categorical
#Producing the one-hot encoded tensors
X_one_hot=to_categorical(X)
y_one_hot=to_categorical(y)
```

```python
model=models.Sequential([
    layers.SimpleRNN(50,input_shape=(seq_length,len(chars)),activation="relu"),
    layers.Dense(len(chars),activation="softmax")
])
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.fit(X_one_hot,y_one_hot,epochs=100,verbose=1)
```

```python
start_seq="I a"
generated_text=start_seq
for i in range(50):
    x = np.array([[char_to_idx[char] for char in generated_text[-seq_length:]]])
    x_one_hot = tf.one_hot(x, len(chars))
    prediction = model.predict(x_one_hot)
    next_index = np.argmax(prediction)
    next_char = idx_to_char[next_index]
    generated_text += next_char

print("Generated Text:")
print(generated_text)
```
